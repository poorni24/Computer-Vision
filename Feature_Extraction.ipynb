```python
import cv2
import numpy as np
import dlib
from google.colab.patches import cv2_imshow


detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("/content/gdrive/MyDrive/shape_predictor_68_face_landmarks.dat")

while True:
    
    #with open('/content/drive/My Drive/Frontal_NonFrontal/frontal.jpg', 'r') as f:
     # frame = f.read()
    frame = cv2.imread('/content/gdrive/MyDrive/facial.jfif')
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = detector(gray)
    for facial in faces:
        x1 = facial.left()
        y1 = facial.top()
        x2 = facial.right()
        y2 = facial.bottom()
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255,255,0), 3)

        landmarks = predictor(gray, facial)

        for n in range(0, 68):
            x = landmarks.part(n).x
            y = landmarks.part(n).y
            cv2.circle(frame, (x, y), 4, (0,0,0), -1)

    cv2_imshow(frame)
    break;
```
